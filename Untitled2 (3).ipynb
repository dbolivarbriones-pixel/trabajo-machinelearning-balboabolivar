# 1. Importamos la base de datos mediante url



```python
#Importamos Base de datos
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print(df.shape)
df.head()
```

# 2. Tratamos los datos ya importados

Primero separamos datos numericos de categorias

```python
#Trabajamos en la categoricacion de la base datos

num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = df.select_dtypes(exclude=[np.number]).columns.tolist()

print("Separamos datos numéricos de los categóricos:")
print("Numéricas:", num_cols)
print("Categóricas:", cat_cols)

# Resumen numérico
df[num_cols].describe().T
```

```python
# Resumen categórico
df[cat_cols].describe().T
```

Identificamos variables con datos atipicos y las tratamos

```python
#Datos atipicos y tratamiento de la base de datos

def tratar_outliers_v2(df, columnas):
    # Hacemos una copia para no modificar el original
    df_procesado = df.copy()
    limites_dict = {}

    for col in columnas:
        # Calculamos los percentiles 1 y 99
        p1, p99 = df_procesado[col].quantile([0.01, 0.99])

        # Guardamos los límites en el diccionario
        limites_dict[col] = (p1, p99)

        # Aplicamos el recorte (clip) directamente
        df_procesado[col] = df_procesado[col].clip(p1, p99)

    return df_procesado, limites_dict

# Aplicación a las columnas numéricas
df_w, bounds = tratar_outliers_v2(df, num_cols)

# Mostrar los límites calculados
print("Límites aplicados (P1 y P99):")
bounds
```

tratamos el passenger para dar muestra a los conocimientos adquiridos, pero realmente este punto no es necesario debido a que el passengerID es un numero identificatorio, no un dato utilmente tratable.

```python
# Comparación antes vs después
col = num_cols[0]

print("Valores extremos originales y procesados:")
print("Antes:", df[col].min(), df[col].max())
print("Después:", df_w[col].min(), df_w[col].max())
```

```python
#Tratamiento de Missing value (Celdas vacias)

# Analizamos la cantidad y porcentaje de valores nulos
resumen_nulos = df_w.isnull().agg(['sum', 'mean']).transpose()

# Renombramos columnas y calculamos el porcentaje
resumen_nulos.columns = ['missing_count', 'missing_pct']
resumen_nulos['missing_pct'] = resumen_nulos['missing_pct'] * 100

# Ordenamos de mayor a menor y mostrar los primeros 15
resumen_nulos = resumen_nulos.sort_values(by='missing_count', ascending=False)
resumen_nulos.head(15)
```

```python
# Creamos una copia del dataframe procesado
df_clean = df_w.copy()

# 1) Imputación de categóricas: reemplazo masivo por "Missing"
df_clean[cat_cols] = df_clean[cat_cols].fillna("Missing")

# 2) Imputación de numéricas: usando la media de cada columna
# Calculamos las medias una sola vez y las aplicamos
medias_columnas = df_clean[num_cols].mean()
df_clean[num_cols] = df_clean[num_cols].fillna(medias_columnas)

# Comprobación final de nulos (estilo resumido)
print("Conteo de nulos tras la limpieza:")
df_clean.isnull().sum().sort_values(ascending=False).head(10)
```

# 3. Analisis de correlacion

```python
# Calculamos la matriz para entender cómo se relacionan nuestras variables numéricas
matriz_corr = df_clean[num_cols].corr(method="pearson")
matriz_corr
```

```python
# Para identificar rápidamente las relaciones más intensas:
# Transformamos a valor absoluto y nos quedamos solo con la parte superior de la matriz
corr_absoluta = matriz_corr.abs()
mascara_superior = np.triu(np.ones(corr_absoluta.shape), k=1).astype(bool)

# Filtramos y ordenamos los pares de variables con mayor correlación
ranking_correlaciones = corr_absoluta.where(mascara_superior).stack().sort_values(ascending=False)

print("Parejas de variables con mayor relación (Top 15):")
ranking_correlaciones.head(15)
```

```python
from sklearn.model_selection import train_test_split

# Definimos nuestra variable objetivo
target = "Survived"

# Separamos las características (X) de la etiqueta que queremos predecir (y)
features = df_clean.drop(columns=[target])
label = df_clean[target]

# Dividimos el dataset: usamos un 20% para validación
X_train, X_val, y_train, y_val = train_test_split(
    features,
    label,
    test_size=0.20,
    random_state=42,
    stratify=label  # Esto asegura que la proporción de sobrevivientes sea igual en ambos grupos
)

# Revisamos las dimensiones resultantes
print(f"Dimensiones de Entrenamiento: {X_train.shape}, {y_train.shape}")
print(f"Dimensiones de Validación: {X_val.shape}, {y_val.shape}")

# Confirmamos que el balance de clases (proporciones) se mantuvo correctamente
print("\nVerificación de proporciones (Target):")
print("Total General:\n", label.value_counts(normalize=True))
print("Grupo de Entrenamiento:\n", y_train.value_counts(normalize=True))
print("Grupo de Validación:\n", y_val.value_counts(normalize=True))
```

# 4. Separamos muestra de entrenamiento y validacion

```python
target = "Survived"
results = []

for col in num_cols:
    if col == target:
        continue

    # Segmentación por clases
    group0 = df_clean[df_clean[target] == 0][col]
    group1 = df_clean[df_clean[target] == 1][col]

    # Test t de Welch
    t_stat, p_value = stats.ttest_ind(group0, group1, equal_var=False)

    # Estadísticos descriptivos y Cohen's D
    mean0 = group0.mean()
    mean1 = group1.mean()
    diff_means = mean1 - mean0

    pooled_std = np.sqrt((group0.var() + group1.var()) / 2)
    cohens_d = diff_means / pooled_std

    results.append({
        "variable": col,
        "mean_class_0": mean0,
        "mean_class_1": mean1,
        "diff_means": diff_means,
        "p_value": p_value,
        "cohens_d": cohens_d
    })

# Creación y ordenamiento del DataFrame de discriminación
df_discrimination = pd.DataFrame(results)

df_discrimination = df_discrimination.sort_values(
    by="cohens_d",
    key=np.abs,
    ascending=False
)

df_discrimination
```

**Introducción**

 En esta segunda etapa del proyecto, daremos continuidad al pipeline reproducible de Machine Learning construido durante la Unidad 1 con el dataset del Titanic. Tras haber superado la fase de exploración, limpieza, imputación de datos faltantes y partición de la muestra, el objetivo central de esta entrega es aplicar y evaluar distintos algoritmos de aprendizaje supervisado.Dado que buscamos predecir la supervivencia de los pasajeros (Sobrevive = 1, No sobrevive = 0), nos enfrentamos a un problema de clasificación binaria. Para resolverlo, entrenaremos modelos paramétricos, de ensamble y basados en redes neuronales, evaluando su capacidad predictiva en las muestras de entrenamiento y validación mediante múltiples métricas de desempeño.

**- Punto 1: Entrenamiento de diferentes modelos de Machine Learning**

Para esta sección, prepararemos los datos convirtiendo las variables categóricas en numéricas (One-Hot Encoding) y entrenaremos los cuatro tipos de modelos solicitados en la rúbrica:

1. Algoritmo de Penalización (ElasticNet)

2. Random Forest

3. XGBoost

4. Red Neuronal Simple (MLPClassifier)

```python
# herramientas nuevas
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# 1. PREPARACIÓN DE DATOS
# Transformamos las variables categóricas (como el Sexo) a columnas numéricas (0 o 1)
# Usamos drop_first=True para evitar multicolinealidad
X_train_encoded = pd.get_dummies(X_train, drop_first=True)
X_val_encoded = pd.get_dummies(X_val, drop_first=True)

# Alineamos las columnas por si alguna categoría estaba en validación pero no en train
X_train_encoded, X_val_encoded = X_train_encoded.align(X_val_encoded, join='left', axis=1, fill_value=0)

# 2. DEFINICIÓN Y ENTRENAMIENTO DE LOS MODELOS

# Modelo 1: Penalización (ElasticNet, que combina penalización L1 y L2)
# Requiere escalar los datos previamente, por lo que usamos un Pipeline
modelo_penalizado = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(penalty="elasticnet", solver="saga", l1_ratio=0.5, max_iter=2000, random_state=42))
])
modelo_penalizado.fit(X_train_encoded, y_train)

# Modelo 2: Random Forest (Modelo de Ensamble tipo Bagging)
modelo_rf = RandomForestClassifier(n_estimators=100, random_state=42)
modelo_rf.fit(X_train_encoded, y_train)

# Modelo 3: XGBoost (Modelo de Ensamble tipo Boosting)
modelo_xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
modelo_xgb.fit(X_train_encoded, y_train)

# Modelo 4: Red Neuronal Simple (Perceptrón Multicapa)
# También requiere escalar los datos para que los gradientes converjan correctamente
modelo_nn = Pipeline([
    ("scaler", StandardScaler()),
    ("model", MLPClassifier(hidden_layer_sizes=(32,), max_iter=1000, random_state=42))
])
modelo_nn.fit(X_train_encoded, y_train)

print("4 modelos entrenados exitosamente con la muestra de entrenamiento!")
```

# Punto 2: Evaluación de Desempeño

Una vez entrenados los modelos, procederemos a evaluar su rendimiento utilizando diversas métricas. Dado que estamos frente a un problema de clasificación binaria con clases algo desbalanceadas, no nos quedaremos solo con la exactitud (Accuracy). Evaluaremos también Precision, Recall, F1-Score, AUC (Área bajo la curva ROC) y el estadístico KS.

Para cumplir con la rúbrica, calcularemos estas métricas tanto en la muestra de entrenamiento como en la de validación para detectar posibles problemas de sobreajuste (overfitting).

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve

# Definimos la función para calcular el estadístico KS
def ks_statistic(y_true, y_prob):
    fpr, tpr, _ = roc_curve(y_true, y_prob)
    return np.max(tpr - fpr)

# Función para extraer todas las métricas solicitadas
def evaluar_metricas(modelo, X_train, y_train, X_val, y_val, nombre):
    # Predicciones de clase y probabilidades
    y_pred_train = modelo.predict(X_train)
    y_prob_train = modelo.predict_proba(X_train)[:, 1]

    y_pred_val = modelo.predict(X_val)
    y_prob_val = modelo.predict_proba(X_val)[:, 1]

    # Retornamos un diccionario con los resultados
    return {
        "Modelo": nombre,
        "Acc_Train": accuracy_score(y_train, y_pred_train),
        "Acc_Val": accuracy_score(y_val, y_pred_val),
        "Prec_Train": precision_score(y_train, y_pred_train),
        "Prec_Val": precision_score(y_val, y_pred_val),
        "Recall_Train": recall_score(y_train, y_pred_train),
        "Recall_Val": recall_score(y_val, y_pred_val),
        "F1_Train": f1_score(y_train, y_pred_train),
        "F1_Val": f1_score(y_val, y_pred_val),
        "AUC_Train": roc_auc_score(y_train, y_prob_train),
        "AUC_Val": roc_auc_score(y_val, y_prob_val),
        "KS_Train": ks_statistic(y_train, y_prob_train),
        "KS_Val": ks_statistic(y_val, y_prob_val)
    }

# Evaluamos los 4 modelos
res_penalizado = evaluar_metricas(modelo_penalizado, X_train_encoded, y_train, X_val_encoded, y_val, "1. ElasticNet")
res_rf = evaluar_metricas(modelo_rf, X_train_encoded, y_train, X_val_encoded, y_val, "2. Random Forest")
res_xgb = evaluar_metricas(modelo_xgb, X_train_encoded, y_train, X_val_encoded, y_val, "3. XGBoost")
res_nn = evaluar_metricas(modelo_nn, X_train_encoded, y_train, X_val_encoded, y_val, "4. Red Neuronal")

# Creamos la tabla y la mostramos ordenada por el F1-Score de validación (la métrica más equilibrada)
tabla_metricas = pd.DataFrame([res_penalizado, res_rf, res_xgb, res_nn])

# Mostramos solo validación para resumir, pero guardamos todas
columnas_val = ["Modelo", "Acc_Val", "Prec_Val", "Recall_Val", "F1_Val", "AUC_Val", "KS_Val"]
tabla_metricas[columnas_val].sort_values(by="F1_Val", ascending=False).round(4)
```

# Punto 3: Optimización de Hiperparámetros

Tras evaluar el desempeño base de nuestros modelos, observamos que los algoritmos basados en ensambles de árboles de decisión (como Random Forest o XGBoost) presentan los mejores resultados en validación. Para mejorar aún más su capacidad de generalización y evitar el sobreajuste, procederemos a optimizar los hiperparámetros del modelo Random Forest.

**Método propuesto: Búsqueda en Cuadrícula con Validación Cruzada (GridSearchCV)**

Este método evalúa exhaustivamente todas las combinaciones posibles dentro de un espacio de parámetros definido ("grilla"). Para cada combinación, el algoritmo realiza una Validación Cruzada (Cross-Validation), dividiendo internamente la muestra de entrenamiento en pliegues (folds) para asegurar que la configuración elegida sea robusta y no dependa de una partición afortunada de los datos.

*   Métrica de éxito: Optimizaremos priorizando el F1-Score, ya que buscamos el mejor equilibrio entre identificar correctamente a los sobrevivientes (Recall) y no equivocarnos al predecirlos (Precision).




```python
from sklearn.model_selection import GridSearchCV

# Ampliamos y ajustamos la grilla para asegurarnos de que la máquina
# encuentre una combinación que mejore el F1-Score en la validación
param_grid_rf = {
    "n_estimators": [50, 100, 150],
    "max_depth": [4, 5, 6, 8, None],
    "min_samples_split": [2, 3, 4],
    "min_samples_leaf": [1, 2]
}

# Configuramos el optimizador
optimizador_rf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_grid=param_grid_rf,
    scoring="f1",
    cv=5,
    n_jobs=-1
)

# Ejecutamos la búsqueda
optimizador_rf.fit(X_train_encoded, y_train)

# Extraemos los mejores hiperparámetros y el modelo optimizado
mejores_parametros = optimizador_rf.best_params_
mejor_rf = optimizador_rf.best_estimator_

print(f"La mejor combinación de hiperparámetros encontrada es: {mejores_parametros}")

# Evaluamos este modelo FINAL optimizado
res_rf_optimizado = evaluar_metricas(mejor_rf, X_train_encoded, y_train, X_val_encoded, y_val, "Random Forest (Optimizado)")

# Mostramos el resultado de validación del modelo final
print("\nDesempeño del Modelo Random Forest Optimizado en Muestra de Validación:")
for metrica, valor in res_rf_optimizado.items():
    if isinstance(valor, float):
        print(f"- {metrica}: {valor:.4f}")
    else:
        print(f"- {metrica}: {valor}")
```

# Punto 4: Interpretación de Resultados y Selección del Modelo Final

**Justificación de la Métrica de Éxito:**

Para este problema de predicción de supervivencia en el Titanic, la precisión absoluta (Accuracy) puede ser engañosa si las clases están desbalanceadas. Por ello, determinamos que la métrica que refleja mejor el éxito del modelo es el F1-Score. Esta métrica nos entrega un balance armónico entre la Precision (de los que predecimos que sobreviven, cuántos realmente lo hicieron) y el Recall (de todos los que realmente sobrevivieron, cuántos logramos identificar). Adicionalmente, observamos el estadístico KS y el AUC para asegurar que el modelo tenga una buena capacidad de separar ambas clases (sobrevivientes vs. fallecidos).

**Selección del Modelo Final:**

Tras evaluar los algoritmos base y aplicar la optimización de hiperparámetros, seleccionamos el Random Forest Optimizado como nuestro modelo final. Este algoritmo demostró ser el más robusto, logrando el mejor equilibrio en la métrica F1-Score en la muestra de validación, superando a los modelos lineales (ElasticNet) y controlando el sobreajuste mejor que la Red Neuronal y el XGBoost base.
