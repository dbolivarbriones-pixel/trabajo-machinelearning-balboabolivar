# 1. Importamos la base de datos mediante url



```python
#Importamos Base de datos
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print(df.shape)
df.head()
```

# 2. Tratamos los datos ya importados

Primero separamos datos numericos de categorias

```python
#Trabajamos en la categoricacion de la base datos

num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = df.select_dtypes(exclude=[np.number]).columns.tolist()

print("Separamos datos numéricos de los categóricos:")
print("Numéricas:", num_cols)
print("Categóricas:", cat_cols)

# Resumen numérico
df[num_cols].describe().T
```

```python
# Resumen categórico
df[cat_cols].describe().T
```

Identificamos variables con datos atipicos y las tratamos

```python
#Datos atipicos y tratamiento de la base de datos

def tratar_outliers_v2(df, columnas):
    # Hacemos una copia para no modificar el original
    df_procesado = df.copy()
    limites_dict = {}

    for col in columnas:
        # Calculamos los percentiles 1 y 99
        p1, p99 = df_procesado[col].quantile([0.01, 0.99])

        # Guardamos los límites en el diccionario
        limites_dict[col] = (p1, p99)

        # Aplicamos el recorte (clip) directamente
        df_procesado[col] = df_procesado[col].clip(p1, p99)

    return df_procesado, limites_dict

# Aplicación a las columnas numéricas
df_w, bounds = tratar_outliers_v2(df, num_cols)

# Mostrar los límites calculados
print("Límites aplicados (P1 y P99):")
bounds
```

tratamos el passenger para dar muestra a los conocimientos adquiridos, pero realmente este punto no es necesario debido a que el passengerID es un numero identificatorio, no un dato utilmente tratable.

```python
# Comparación antes vs después
col = num_cols[0]

print("Valores extremos originales y procesados:")
print("Antes:", df[col].min(), df[col].max())
print("Después:", df_w[col].min(), df_w[col].max())
```

```python
#Tratamiento de Missing value (Celdas vacias)

# Analizamos la cantidad y porcentaje de valores nulos
resumen_nulos = df_w.isnull().agg(['sum', 'mean']).transpose()

# Renombramos columnas y calculamos el porcentaje
resumen_nulos.columns = ['missing_count', 'missing_pct']
resumen_nulos['missing_pct'] = resumen_nulos['missing_pct'] * 100

# Ordenamos de mayor a menor y mostrar los primeros 15
resumen_nulos = resumen_nulos.sort_values(by='missing_count', ascending=False)
resumen_nulos.head(15)
```

```python
# Creamos una copia del dataframe procesado
df_clean = df_w.copy()

# 1) Imputación de categóricas: reemplazo masivo por "Missing"
df_clean[cat_cols] = df_clean[cat_cols].fillna("Missing")

# 2) Imputación de numéricas: usando la media de cada columna
# Calculamos las medias una sola vez y las aplicamos
medias_columnas = df_clean[num_cols].mean()
df_clean[num_cols] = df_clean[num_cols].fillna(medias_columnas)

# Comprobación final de nulos (estilo resumido)
print("Conteo de nulos tras la limpieza:")
df_clean.isnull().sum().sort_values(ascending=False).head(10)
```

# 3. Analisis de correlacion

```python
# Calculamos la matriz para entender cómo se relacionan nuestras variables numéricas
matriz_corr = df_clean[num_cols].corr(method="pearson")
matriz_corr
```

```python
# Para identificar rápidamente las relaciones más intensas:
# Transformamos a valor absoluto y nos quedamos solo con la parte superior de la matriz
corr_absoluta = matriz_corr.abs()
mascara_superior = np.triu(np.ones(corr_absoluta.shape), k=1).astype(bool)

# Filtramos y ordenamos los pares de variables con mayor correlación
ranking_correlaciones = corr_absoluta.where(mascara_superior).stack().sort_values(ascending=False)

print("Parejas de variables con mayor relación (Top 15):")
ranking_correlaciones.head(15)
```

```python
from sklearn.model_selection import train_test_split

# Definimos nuestra variable objetivo
target = "Survived"

# Separamos las características (X) de la etiqueta que queremos predecir (y)
features = df_clean.drop(columns=[target])
label = df_clean[target]

# Dividimos el dataset: usamos un 20% para validación
X_train, X_val, y_train, y_val = train_test_split(
    features,
    label,
    test_size=0.20,
    random_state=42,
    stratify=label  # Esto asegura que la proporción de sobrevivientes sea igual en ambos grupos
)

# Revisamos las dimensiones resultantes
print(f"Dimensiones de Entrenamiento: {X_train.shape}, {y_train.shape}")
print(f"Dimensiones de Validación: {X_val.shape}, {y_val.shape}")

# Confirmamos que el balance de clases (proporciones) se mantuvo correctamente
print("\nVerificación de proporciones (Target):")
print("Total General:\n", label.value_counts(normalize=True))
print("Grupo de Entrenamiento:\n", y_train.value_counts(normalize=True))
print("Grupo de Validación:\n", y_val.value_counts(normalize=True))
```

# 4. Separamos muestra de entrenamiento y validacion

```python
target = "Survived"
results = []

for col in num_cols:
    if col == target:
        continue

    # Segmentación por clases
    group0 = df_clean[df_clean[target] == 0][col]
    group1 = df_clean[df_clean[target] == 1][col]

    # Test t de Welch
    t_stat, p_value = stats.ttest_ind(group0, group1, equal_var=False)

    # Estadísticos descriptivos y Cohen's D
    mean0 = group0.mean()
    mean1 = group1.mean()
    diff_means = mean1 - mean0

    pooled_std = np.sqrt((group0.var() + group1.var()) / 2)
    cohens_d = diff_means / pooled_std

    results.append({
        "variable": col,
        "mean_class_0": mean0,
        "mean_class_1": mean1,
        "diff_means": diff_means,
        "p_value": p_value,
        "cohens_d": cohens_d
    })

# Creación y ordenamiento del DataFrame de discriminación
df_discrimination = pd.DataFrame(results)

df_discrimination = df_discrimination.sort_values(
    by="cohens_d",
    key=np.abs,
    ascending=False
)

df_discrimination
```
